```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
data = {'Hours': np.random.rand(100) * 10}
df = pd.DataFrame(data)
df['Scores'] = df['Hours'] * 10 + np.random.randn(100) * 5
```

```python
X = df[['Hours']]
y = df['Scores']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plt.scatter(X, y, color='black', label='Actual Scores')
plt.plot(X, model.predict(X), color='blue', linewidth=3, label='Regression Line')
plt.xlabel('Hours of Study')
plt.ylabel('Scores')
plt.title('Student Performance Prediction with Regression Line')
plt.legend()
plt.show()
```

```python
print("R2 Score:", metrics.r2_score(y_test, y_pred))
```
